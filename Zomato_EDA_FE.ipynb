```python
one_hot_encoded_data = pd.get_dummies(df2, columns = ['Has Table booking','Has Online delivery','Is delivering now','Switch to order menu','Price range'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data.drop(['Restaurant ID','Restaurant Name','Country Code','City','Address','Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Cuisines', 'Currency', 'Rating color' , 'Rating text' , 'Votes', 'Country'], axis=1, inplace=True)
one_hot_encoded_data.head()
```

```python
one_hot_encoded_data.drop(['Has Table booking_No','Has Online delivery_No', 'Is delivering now_No','Switch to order menu_No'], axis=1, inplace=True)
one_hot_encoded_data.head()
```

```python
X = one_hot_encoded_data[['Has Table booking_Yes', 'Has Online delivery_Yes', 'Is delivering now_Yes', 'Price range_1','Price range_2','Price range_3','Price range_4','Average Cost for two']]
X
```

```python
y = one_hot_encoded_data[['Aggregate rating']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_train, y_train)
```

```python
reg.predict(x_test)
```
